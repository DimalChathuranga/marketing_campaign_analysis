```python
import pandas as pd

# Separate numerical and categorical data
numerical_df = df.select_dtypes(include=['number'])
categorical_df = df.select_dtypes(include=['object', 'category'])
```

```python
#Using Z-Score
from scipy.stats import zscore
import numpy as np

# Z-score method
z_scores = np.abs(zscore(numerical_df))
numerical_outliers = numerical_df[(z_scores > 3).any(axis=1)]
print(numerical_outliers)
```

```python
#iqr outliers detection
def detect_iqr_outliers(dataframe):
    outlier_mask = pd.DataFrame(index=dataframe.index)
    for column in dataframe.columns:
        Q1 = dataframe[column].quantile(0.25)
        Q3 = dataframe[column].quantile(0.75)
        IQR = Q3 - Q1
        outlier_condition = (dataframe[column] < (Q1 - 1.5 * IQR)) | (dataframe[column] > (Q3 + 1.5 * IQR))
        outlier_mask[column] = outlier_condition
    return outlier_mask

numerical_outlier_mask = detect_iqr_outliers(numerical_df)
print(numerical_outliers)
```

```python
# Calculate the IQR for the Income column
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

# Identify the outliers in the Income column
outliers = df[(df['Income'] < (Q1 - 1.5 * IQR)) | (df['Income'] > (Q3 + 1.5 * IQR))]

# Print the number of outliers
print("Number of outliers in the Income column:", len(outliers))
```

Feature engineering-Create new features

```python
print("Unique values in Education column:", df["Education"].unique())
print("Unique values in Martial Status column:", df["Marital_Status"].unique())
```

```python
def education_level(education):
    if education in ["Graduation","phD","Master"]:
        return "high"
    elif education in ["Basic"]:
        return "Middle"
    else:
        return "Low"

df["Education_level"] = df["Education"].apply(education_level)
```

```python
df
```

```python
def living_status(marital_status):
    if marital_status in ['Alone', 'Absurd', 'YOLO']:
        return 'Living Alone'
    else:
        return 'Living with Others'

df['Living_Status'] = df['Marital_Status'].apply(living_status)
```

```python
df_model=df.copy()
df_model
```

Categorical Values convert to numerical values 

```python
df_encoded = pd.get_dummies(df_model, columns=['Education','Marital_Status','Education_level','Living_Status'])
```

```python
nan=df_encoded.isna().sum()
print(nan)
```

```python
df_encoded=df_encoded.fillna(df_encoded.mean())
```

```python
nan=df_encoded.isna().sum()
print(nan)
```

Standardization 

```python
from sklearn.preprocessing import StandardScaler
# Replace infinity with NaN
df_encoded.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill NaN with mean or median
df_encoded.fillna(df_encoded.mean(), inplace=True)
# Initialize StandardScaler
scaler = StandardScaler()

# Apply Z-score standardization to normalize the data
df_standardized = pd.DataFrame(scaler.fit_transform(df_encoded), columns=df_encoded.columns)
```

```python
df_standardized
```

```python
x=df_standardized.drop(columns='Income')
y=df_standardized['Income']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```

Train Linear Regression model

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = np.round(model.predict(X_test))
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Get coefficients and feature names
coefficients = model.coef_
feature_names = X_train.columns

# Create a DataFrame to store coefficients and their corresponding feature names
coefficients_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})

# Display coefficients
print("Coefficients:")
coefficients_df
```

```python
# Plotting coefficients
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
x_pos = np.arange(len(coefficients_df))
plt.bar(x_pos, coefficients_df['Coefficient'], align='center', alpha=0.7)
plt.xticks(x_pos, coefficients_df['Feature'], rotation='vertical')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.title('Feature Coefficients (Linear Regression)')
plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```
